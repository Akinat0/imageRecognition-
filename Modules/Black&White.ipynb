# Import

```python
import cv2
from matplotlib import pyplot as plt

im_gray = cv2.imread('/Users/amir/Desktop/ImageRecognition/Images/360pans/cubemap.jpg', 0)
im_part1 = cv2.imread('/Users/amir/Desktop/ImageRecognition/Images/360pans/Pieces/IMG_1274.jpg', 0)

image = im_part1

plt.figure(figsize=(10, 10))
plt.imshow(image, 'gray'),plt.show()

```

# Normalization


```python
#im_gray = cv2.normalize(im_gray, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
plt.figure(figsize=(10, 10))
plt.imshow(image, 'gray'),plt.show()
```

# Check the difference

```python
plt.figure(figsize=(10, 10))
plt.imshow(image-im_gray, 'gray'),plt.show()
```

# Blur

```python
image = cv2.medianBlur(image,5)
plt.figure(figsize=(10, 10))
plt.imshow(image, 'gray'),plt.show()
```

```python

(thresh, image) = cv2.threshold(image, 60, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)

plt.figure(figsize=(100, 100))
plt.imshow(image, 'gray'),plt.show()
```
