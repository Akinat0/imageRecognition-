# Import

```python
import cv2
from matplotlib import pyplot as plt

im_gray = cv2.imread('/Users/amir/Desktop/ImageRecognition/Images/360pans/cubemap.jpg', 0)

image = im_gray

plt.figure(figsize=(10, 10))
plt.imshow(image, 'gray'),plt.show()

```

# Normalization


```python
image = cv2.normalize(image, None, alpha=0, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
plt.figure(figsize=(10, 10))
plt.imshow(image, 'gray'),plt.show()
```

# Check the difference

```python
plt.figure(figsize=(10, 10))
plt.imshow(image-im_gray, 'gray'),plt.show()
```

```python
image = cv2.medianBlur(image,5)
plt.figure(figsize=(10, 10))
plt.imshow(image, 'gray'),plt.show()
```

```python
#image = cv2.medianBlur(image,5)
im_bw = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
#thresh = 127
#im_bw = cv2.threshold(im_gray, 65, 255, cv2.THRESH_BINARY)[1]

plt.figure(figsize=(100, 100))
plt.imshow(im_bw, 'gray'),plt.show()
```
